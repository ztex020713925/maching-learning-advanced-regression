```python
import pandas as pd
train_df = pd.read_csv("train.csv", encoding="utf-8")
test_df = pd.read_csv("test.csv", encoding="utf-8")
```

```python
train_df
# 1460筆
```

```python
# 把兩個表格合併在一起，一起處理

total_df = pd.concat([train_df, test_df], axis=0)
total_df

# 可以看到後面"SalePrice"有缺，要把它填成0
```

```python
# data 把答案丟掉
data = total_df.drop(["SalePrice"], axis=1)

# 這是答案
y_train = train_df["SalePrice"]
y_train
```

```python
# 補缺失
# 先看
data.isna().sum()
```

```python
# 看中位數
# MSSubClass 這個不能補，要丟掉

med = data.median().drop(["MSSubClass"])
data.median()
```

```python
# 把缺的值都填成中位數，這邊只補數字類型的

data = data.fillna(med)
```

```python
# 剩下的應該都是文字類型了

s = data.isna().sum()
s[s>0]
```

```python
# One-Hot encoding
# 把文字類型的都變成是非題

pd.get_dummies(data)
```

```python
data = pd.get_dummies(data)


data = pd.get_dummies(data, columns=["MSSubClass"])
# 丟掉"Id"
data = data.drop(["Id"], axis=1)
data
```

```python
# 檢查還有沒有缺欄位

s = data.isna().sum()
s[s >0]
```

```python
# 先補完 最後才用到, 分開x_train, x_test
train_df.shape
# loc: 列標籤, iloc: 0, 1...
# iloc -> ["第一筆", "第二筆"....]


x_train = data.iloc[:train_df.shape[0]]
x_test = data.iloc[train_df.shape[0]:]
x_test
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score

# 找參數~~慢慢調
# 會找很久，建議一開始範圍選大一點
# params = {
#     "n_estimators":range(20, 150, 10) , 
#     "max_depth":range(5, 20)
# }

# reg = RandomForestRegressor()
# cv = GridSearchCV(reg, 
#                   params, 
#                   cv=10, 
#                   n_jobs=4, 
#                   scoring="r2")

# cv.fit(x_train, y_train)

# print(cv.best_params_)
# print(cv.best_score_)
```

```python
# 調整到好參數之後就在下面套用，準備猜答案

reg = RandomForestRegressor(n_estimators=75, max_depth=8)
reg.fit(x_train, y_train)
pre = reg.predict(x_test)
test_id = test_df["Id"]
result = pd.DataFrame({
    "Id":test_id,
    "SalePrice":pre
})
result.to_csv("house_rf.csv", 
              encoding="utf-8",
              index=False)
result
```
